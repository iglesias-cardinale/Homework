```python
import numpy as np 
import matplotlib.pyplot as plt
```

The expression for calculating the Madelung Constant for the rock-salt structure is $$\alpha = \sum_{j \neq i} \pm \frac{1}{P_{ij}} = \sum_{m,n,l = -\infty}^\infty \frac{\left(-1\right)^{m+n+l}}{\left(m^2+n^2+l^2\right)^{1/2}}$$

```python
## Performing the sum

#Define the term in the sum as a function

def madelung_term(m, n, l):
    num = (-1)**(m+n+l)
    den = (m**2 + n**2 + l**2)**0.5

    return num/den

# Set ranges for m, n and l
range_max = 70

m_list = np.array(range(-range_max, range_max), dtype = float)
n_list = np.array(range(-range_max, range_max), dtype = float)
l_list = np.array(range(-range_max, range_max), dtype = float)


# Set initial constant to which we will add
madelung = 0

for m in m_list:
    for n in n_list:
        for l in l_list:
            if m == n == l == 0:
                madelung += 0
            else:
                madelung += madelung_term(m,n,l)

print(madelung)
```

```python
## Make it a function so we can plot

def madelung_func(range_max):
    m_list = np.array(range(-range_max, range_max), dtype = float)
    n_list = np.array(range(-range_max, range_max), dtype = float)
    l_list = np.array(range(-range_max, range_max), dtype = float)

    # Set initial constant to which we will add
    madelung = 0

    for m in m_list:
        for n in n_list:
            for l in l_list:
                if m == n == l == 0:
                    madelung += 0
                else:
                    madelung += madelung_term(m,n,l)

    return madelung 
```

```python
# Generate data for plotting
num_range = np.arange(1, 50, dtype=int)  # Ensure num_range contains integers
madelung_list = [madelung_func(r) for r in num_range]

plt.plot(num_range, madelung_list, marker='o', linestyle='-', color='b')
plt.xlabel('Range Max')
plt.ylabel('Madelung Constant')
plt.title('Convergence of Madelung Constant with Range Max')
plt.grid()
plt.show()
```
