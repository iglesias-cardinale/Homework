```python
import numpy as np
import matplotlib.pyplot as plt

# Load data from 'band.dat'
data = np.loadtxt("band.dat")

# Separate K values (first column) and eigenvalues (remaining columns)
K_values = data[:, 0]
eigenvalues = data[:, 1:]

# Plot each band (each eigenvalue column) as a function of K
plt.figure(figsize=(10, 6))
for band in eigenvalues.T:
    plt.plot(K_values, band, label="Band")

# Labeling the plot
plt.xlabel("K (Wavevector)")
plt.ylabel("Energy (a.u.)")
plt.title("Band Structure")
plt.grid(True)
plt.tight_layout()
plt.show()
```
